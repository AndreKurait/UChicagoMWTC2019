```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
def load_object(file_name):
    "load the pickled object"
    with open(file_name, 'rb') as f:
        return pickle.load(f)
```

```python
#Data Loading
#vals = ['k','m', 'n', 'q', 'u', 'v']
k_array = load_object('k_array.pkl')
k_array = np.stack(k_array)
m_array = load_object('m_array.pkl')
m_array = np.stack(m_array)
n_array = load_object('n_array.pkl')
n_array = np.stack(n_array)
q_array = load_object('q_array.pkl')
q_array = np.stack(q_array)
u_array = load_object('u_array.pkl')
u_array = np.stack(u_array)
v_array = load_object('v_array.pkl')
v_array = np.stack(v_array)
```

```python
'''
K - May
M - June
N - July
Q - Aug
U - Sept
V - Oct
'''
```
